# Preguntas


### - ¿Se aumentó el nivel escolar en los estados con menor grado de escolaridad a lo largo de los años?
### - ¿Qué estados tienen el mayor grado de escolaridad en el país? ¿Se mantuvo a lo largo de los años?
### - ¿En qué estados hay más desigualdad entre la escolaridad de ambos sexos?
### - ¿Qué estados han tenido el mayor crecimiento en grado de escolaridad?
### - ¿Cuál es el grado de escolaridad de Guanajuato respecto al promedio del país?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

font = {'family': 'serif',
        'color':  'black',
        'weight': 'normal',
        'size': 16,
        }

years = np.array([2000, 2010, 2015])
index = np.array(list(range(0, 5)))

edReport = pd.read_excel('educacion.xlsx')
edReport.rename(columns={'Unnamed: 0': 'Estado', 'Unnamed 1': '2000/Total', 'Unnamed: 2': '2000/H', 'Unnamed: 3': '2000/M',
                         'Unnamed: 4': '2010/Total', 'Unnamed: 5': '2010/H', 'Unnamed: 6': '2010/M',
                         'Unnamed: 8': '2015/H', 'Unnamed: 9': '2015/M'}, inplace=True)
edReport = edReport.set_index('Estado')
edReport.rename(index={'Michoacán de Ocampo': 'Michoacán', 'Veracruz de Ignacio de la Llave': 'Veracruz'}, inplace=True)
edReport.drop(np.nan, inplace = True)
edReport
```

```python
edReport.sort_values(by = '2000', ascending = True, inplace = True)
edReport_lowest2000 = edReport[:5] #Estaría chido añadir la row de Estados Unidos Mexicanos para comparar con el promedio
edReport_lowest2000 = edReport_lowest2000.drop(['2010', '2010/H', '2010/M', '2015', '2015/H', '2015/M'], axis = 1)
edReport_lowest2000.rename(columns = {'2000': 'Total', '2000/H': 'Hombres', '2000/M': 'Mujeres'}, inplace = True)

edReport.sort_values(by = '2000', ascending = False, inplace = True)
edReport_highest2000 = edReport[:5]
edReport_highest2000 = edReport_highest2000.drop(['2010', '2010/H', '2010/M', '2015', '2015/H', '2015/M'], axis = 1)
edReport_highest2000.rename(columns = {'2000': 'Total', '2000/H': 'Hombres', '2000/M': 'Mujeres'}, inplace = True)

edReport.sort_values(by = '2010', ascending = True, inplace = True)
edReport_lowest2010 = edReport[:5] 
edReport_lowest2010 = edReport_lowest2010.drop(['2000', '2000/H', '2000/M', '2015', '2015/H', '2015/M'], axis = 1)
edReport_lowest2010.rename(columns = {'2010': 'Total', '2010/H': 'Hombres', '2010/M': 'Mujeres'}, inplace = True)

edReport.sort_values(by = '2010', ascending = False, inplace = True)
edReport_highest2010 = edReport[:5]
edReport_highest2010 = edReport_highest2010.drop(['2000', '2000/H', '2000/M', '2015', '2015/H', '2015/M'], axis = 1)
edReport_highest2010.rename(columns = {'2010': 'Total', '2010/H': 'Hombres', '2010/M': 'Mujeres'}, inplace = True)

edReport.sort_values(by = '2015', ascending = True, inplace = True)
edReport_lowest2015 = edReport[:5] 
edReport_lowest2015 = edReport_lowest2015.drop(['2000', '2000/H', '2000/M', '2010', '2010/H', '2010/M'], axis = 1)
edReport_lowest2015.rename(columns = {'2015': 'Total', '2015/H': 'Hombres', '2015/M': 'Mujeres'}, inplace = True)

edReport.sort_values(by = '2015', ascending = False, inplace = True)
edReport_highest2015 = edReport[:5]
edReport_highest2015 = edReport_highest2015.drop(['2000', '2000/H', '2000/M', '2010', '2010/H', '2010/M'], axis = 1)
edReport_highest2015.rename(columns = {'2015': 'Total', '2015/H': 'Hombres', '2015/M': 'Mujeres'}, inplace = True)
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

width = 0.15
plt.bar(index, edReport_highest2000.Total, width = width, label = '2000', alpha = 0.7)
plt.bar(index + width, edReport_highest2010.Total, width = width, label = '2010', alpha = 0.7)
plt.bar(index + 2*width, edReport_highest2015.Total, width = width, label = '2015', alpha = 0.7)
plt.hlines(edReport.iloc[0]['2000'], index[0] - 0.25, index[-1] + 0.5, color = 'blue', linestyle = 'dotted', label = 'México 2000')
plt.hlines(edReport.iloc[0]['2010'], index[0] - 0.25, index[-1] + 0.5, color = 'orange', linestyle = 'dotted', label = 'México 2010')
plt.hlines(edReport.iloc[0]['2015'], index[0] - 0.25, index[-1] + 0.5, color = 'green', linestyle = 'dotted', label = 'México 2015')

ax.set_xlim(index[0] - 0.25, index[-1] + 0.5)

plt.xlabel('Estados', fontsize = 16)
plt.ylabel('Grado de Escolaridad', fontsize = 16)
plt.title('Estados con el Mayor Grado de Escolaridad', fontsize = 16)
plt.xticks(index + 0.8*width, edReport_highest2000.index)
plt.legend(loc = 'lower right')
plt.grid(color='#95a5a6', linestyle='--', linewidth=2, axis='y', alpha=0.3)
plt.savefig('MiniProyecto\mayorGrado.png' , bbox_inches='tight')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

width = 0.15
plt.bar(index, edReport_lowest2000.Total, width = width, label = '2000', alpha = 0.7)
plt.bar(index + width, edReport_lowest2010.Total, width = width, label = '2010', alpha = 0.7)
plt.bar(index + 2*width, edReport_lowest2015.Total, width = width, label = '2015', alpha = 0.7)
plt.hlines(edReport.iloc[0]['2000'], index[0] - 0.25, index[-1] + 0.5, color = 'blue', linestyle = 'dotted', label = 'México 2000')
plt.hlines(edReport.iloc[0]['2010'], index[0] - 0.25, index[-1] + 0.5, color = 'orange', linestyle = 'dotted', label = 'México 2010')
plt.hlines(edReport.iloc[0]['2015'], index[0] - 0.25, index[-1] + 0.5, color = 'green', linestyle = 'dotted', label = 'México 2015')

ax.set_xlim(index[0] - 0.25, index[-1] + 0.5)

plt.xlabel('Estados', fontsize = 16)
plt.ylabel('Grado de Escolaridad', fontsize = 16)
plt.title('Estados con el Menor Grado de Escolaridad', fontsize = 16)
plt.xticks(index + 0.8*width, edReport_lowest2000.index)
plt.legend(loc = 'center right')
plt.grid(color='#95a5a6', linestyle='--', linewidth=2, axis='y', alpha=0.3)
plt.savefig('MiniProyecto\menorGrado.png' , bbox_inches='tight')
plt.show()
```

```python
edReport['Diferencia 2000'] = np.abs(edReport['2000/H'] - edReport['2000/M'])
edReport['Diferencia 2010'] = np.abs(edReport['2010/H'] - edReport['2010/M'])
edReport['Diferencia 2015'] = np.abs(edReport['2015/H'] - edReport['2015/M'])

edReport.sort_values(by = 'Diferencia 2000', ascending = False, inplace = True)
edReport_dif2000 = edReport[:5]
edReport_dif2000 = edReport_dif2000.drop(['2000', '2000/H', '2000/M',
                                          '2010', '2010/H', '2010/M',
                                          '2015', '2015/H', '2015/M',
                                          'Diferencia 2010', 'Diferencia 2015'], axis = 1)
edReport_dif2000.rename(columns = {'Diferencia 2000': 'Diferencia'}, inplace = True)

edReport.sort_values(by = 'Diferencia 2010', ascending = False, inplace = True)
edReport_dif2010 = edReport[:5]
edReport_dif2010 = edReport_dif2010.drop(['2000', '2000/H', '2000/M',
                                          '2010', '2010/H', '2010/M',
                                          '2015', '2015/H', '2015/M',
                                          'Diferencia 2000', 'Diferencia 2015'], axis = 1)
edReport_dif2010.rename(columns = {'Diferencia 2010': 'Diferencia'}, inplace = True)


edReport.sort_values(by = 'Diferencia 2015', ascending = False, inplace = True)
edReport_dif2015 = edReport[:5]
edReport_dif2015 = edReport_dif2015.drop(['2000', '2000/H', '2000/M',
                                          '2010', '2010/H', '2010/M',
                                          '2015', '2015/H', '2015/M',
                                          'Diferencia 2000', 'Diferencia 2010'], axis = 1)
edReport_dif2015.rename(columns = {'Diferencia 2015': 'Diferencia'}, inplace = True)
```

Decidí tomar los primeros 5 estados en lugar de los primeros 3 puesto que al graficar los primeros 3 las posiciones no cambiaban a lo largo de los años, sólo se reducía la diferencia en el 2015.

```python
edReport['Incremento'] = edReport['2015'] - edReport['2000']
edReport.sort_values(by = 'Incremento', ascending = False, inplace = True)
edReport_increase = edReport[:5]
edReport_increase = edReport_increase.rename(columns = {'2000': 'Total2000', '2010': 'Total2010', '2015': 'Total2015'})
edReport_increase
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

width = 0.15
plt.bar(index, edReport_increase.Total2000, width = width, label = '2000', color = 'blue', alpha = 0.5)
plt.bar(index + width, edReport_increase.Total2010, width = width, label = '2010', color = 'purple', alpha = 0.5)
plt.bar(index + 2*width, edReport_increase.Total2015, width = width, label = '2015', color = 'yellow', alpha = 0.5)

plt.xlabel('Estados', fontsize = 16)
plt.ylabel('Grado de Escolaridad', fontsize = 16)
plt.title('Estados con el Mayor Crecimiento Educativo', fontsize = 16)
plt.xticks(index + 0.8*width, edReport_increase.index)
plt.legend(loc = 'lower right')
plt.grid(color='#95a5a6', linestyle='--', linewidth=2, axis='y', alpha=0.5)
plt.text(0.095, 0.5, '2.1', fontdict = font, horizontalalignment='center', verticalalignment='center', transform=ax.transAxes)
plt.text(0.297, 0.5, '2.1', fontdict = font, horizontalalignment='center', verticalalignment='center', transform=ax.transAxes)
plt.text(0.502, 0.5, '2.1', fontdict = font, horizontalalignment='center', verticalalignment='center', transform=ax.transAxes)
plt.text(0.705, 0.5, '2.0', fontdict = font, horizontalalignment='center', verticalalignment='center', transform=ax.transAxes)
plt.text(0.910, 0.5, '2.0', fontdict = font, horizontalalignment='center', verticalalignment='center', transform=ax.transAxes)
plt.savefig('MiniProyecto\crecimiento.png' , bbox_inches='tight')
plt.show()
```

```python
edReport.sort_values(by = 'Estado', ascending = True, inplace = True)
edReportGto = edReport[edReport.index == 'Guanajuato']
edReportMex = edReport[edReport.index == 'Estados Unidos Mexicanos']

edReportGto = pd.concat([edReportGto, edReportMex])
edReportGto.rename(columns={'2000/H': 'Hombres Año 2000', '2000/M': 'Mujeres Año 2000',
                            '2010/H': 'Hombres Año 2010', '2010/M': 'Mujeres Año 2010',
                            '2015/H': 'Hombres Año 2015', '2015/M': 'Mujeres Año 2015'}, inplace = True)
edReportGto.drop(['2000', '2010', '2015', 'Diferencia 2000', 'Diferencia 2010', 'Diferencia 2015', 'Incremento'], axis = 1, inplace = True)
edReportGtoH = edReportGto
edReportGtoM = edReportGto
edReportGtoH = edReportGtoH.drop(['Mujeres Año 2000', 'Mujeres Año 2010', 'Mujeres Año 2015'], axis = 1)
edReportGtoM = edReportGtoM.drop(['Hombres Año 2000', 'Hombres Año 2010', 'Hombres Año 2015'], axis = 1)
edReportGtoH
```
