# Método de Euler (código tipo C)

Se presenta la resolución de la ecuación del movimiento para un oscilador armónico utilizando el método de Euler. La ecuación a resolver es $\ddot{\theta} + \omega^2 \theta = 0$. Primero se evitó el uso de la librería numpy, utilizando el siguiente código:

```python
import numpy as np
import matplotlib.pyplot as plt
import timeit

#Para obtener el tiempo de ejecución
start = timeit.default_timer()

#Paso y número máximo de iteraciones
NMAX = 10000
h = 0.001

#Condiciones iniciales y parámetros
omega = 1
theta0 = 0.1
thetaDot0 = 0

#Declaración de las listas y subsecuente llenado con ceros para posteriormente poder cambiar cada elemento
tValues = []
thetaValues = []
thetaDotValues = []

for i in range(0, NMAX):
    tValues.append(i/1000)
    thetaValues.append(0)
    thetaDotValues.append(0)
    
thetaValues[0] = theta0
thetaDotValues[0] = thetaDot0

#Ejecución del método
#Iniciando en 1 para no utilizar i + 1 como índice y evitar errores al llegar a NMAX
for i in range(1, NMAX):
    thetaValues[i] = thetaValues[i - 1] + h * thetaDotValues[i - 1]
    thetaDotValues[i] = thetaDotValues[i - 1] - h * omega**2 * thetaValues[i - 1]
    

plt.plot(tValues, thetaValues, label = '$\Theta$')
plt.plot(tValues, thetaDotValues, label = '$\dot{\Theta}$')
plt.xlabel('Tiempo (s)')
plt.ylabel('Ángulo $\Theta$ (rad) / Velocidad $\dot{\Theta}$ (rad/s)')
plt.grid()
plt.title('Movimiento Armónico Simple')
plt.legend()
plt.show()

stop = timeit.default_timer()
executionTime = stop - start

print("Tiempo de ejecución: " + str(executionTime))
```

# Método de Euler (NumPy)

Utilizando la librería numpy, se cambió la forma de llenar los arreglos utilizando los métodos _linspace_ y _zeros_.

```python
import numpy as np
import matplotlib.pyplot as plt
import timeit

#Para obtener el tiempo de ejecución
start = timeit.default_timer()

#Paso y número máximo de iteraciones
NMAX = 10000
h = 0.001

#Condiciones iniciales y parámetros 
omega = 1
theta0 = 0.1
thetaDot0 = 0

#Declaración de las listas y llenado con ceros (usando numpy) para posteriormente poder cambiar cada elemento
tValues = np.linspace(0, 10, NMAX)
thetaValues = np.zeros(NMAX)
thetaDotValues = np.zeros(NMAX)
    
thetaValues[0] = theta0
thetaDotValues[0] = thetaDot0

#Ejecución del método
#Iniciando en 1 para no utilizar i + 1 como índice y evitar errores al llegar a NMAX.
for i in range(1, NMAX):
    thetaValues[i] = thetaValues[i - 1] + h * thetaDotValues[i - 1]
    thetaDotValues[i] = thetaDotValues[i - 1] - h * omega**2 * thetaValues[i - 1]
    
plt.plot(tValues, thetaValues, label = '$\Theta$')
plt.plot(tValues, thetaDotValues, label = '$\dot{\Theta}$')
plt.xlabel('Tiempo (s)')
plt.ylabel('Ángulo $\Theta$ (rad) / Velocidad $\dot{\Theta}$ (rad/s)')
plt.grid()
plt.title('Movimiento Armónico Simple')
plt.legend()
plt.show()

stop = timeit.default_timer()
executionTime = stop - start

print("Tiempo de ejecución: " + str(executionTime))
```
