```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
from scipy import special as sp

mu = 70
sigma = 2
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
NMAX = 10000
radius = 5

xValues, yValues = [], []

for i in range(0, NMAX):
    x = np.random.rand(1) * 2*radius - radius
    y = np.random.rand(1) * 2*radius - radius   
    if x**2 + y** 2 <= radius**2:
        xValues.append(x)
        yValues.append(y)

plt.plot(xValues, yValues, 'r.', marker = 'x')
plt.title('Crculo de Radio 5', fontsize = 16)
plt.xlabel('X', fontsize = 14)
plt.ylabel('Y', fontsize = 14)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
NMAX = 10000
radius = 5

xValues, yValues, zValues = [], [], []

for i in range(0, NMAX):
    x = np.random.rand(1) * 2*radius - radius
    y = np.random.rand(1) * 2*radius - radius
    z = np.random.rand(1) * 2*radius - radius
    if x**2 + y**2 + z**2 <= radius**2:
        xValues.append(x)
        yValues.append(y)
        zValues.append(z)

plt.title('Esfera de Radio 5', fontsize = 16)
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('X axis', fontsize = 14)
ax.set_ylabel('Y axis', fontsize = 14)
ax.set_zlabel('Z axis', fontsize = 14)
ax.scatter(xValues, yValues, zValues, "r", marker='x', color = 'red')
```

```python
NMAX = 100000

def cdfNormal(x, mu, sigma):
    return (1/2)*(1 + sp.erf((x - mu)/(np.sqrt(2)*sigma)))

def cdfNormalInverse(y, mu, sigma):
    return sigma * np.sqrt(2) * sp.erfinv(2*y - 1) + mu

def normal(x, mu, sigma):
    return (1/(sigma * np.sqrt(2*np.pi))) * np.exp(-0.5 * ((x - mu)/sigma)**2)

xValues = np.linspace(mu - 5, mu + 5, NMAX)
yValues = cdfNormal(xValues, mu, sigma)

plt.figure(figsize = [10, 8])
plt.title('Normal CDF ($\mu$ = 70, $\sigma$ = 2)', fontsize = 14)
plt.xlabel('$H_0$', fontsize = 14)
plt.ylabel('F($H_0$)', fontsize = 14)
plt.plot(xValues, yValues, color = 'purple', linestyle = 'dashed')
plt.show()
```

```python
valuesTaken = 50000
h0Values = []

np.random.seed()

for i in range(0, valuesTaken):
    h0Values.append(cdfNormalInverse(float(np.random.rand(1)), mu, sigma))


plt.figure(figsize = [10, 8])
plt.hist(h0Values, bins = 50, alpha = 0.5, histtype = 'stepfilled', edgecolor = 'black')
plt.plot(xValues, 18000*normal(xValues, mu, sigma), color = 'blue')
plt.xlabel('$H_0$', fontsize = 14)
plt.ylabel('Frecuencia', fontsize = 14)
plt.title('CDF Inversa (50,000 elementos)')
plt.show()
```
